```python
num1 = 2
num2 = 4
if num1 > num2:
    print("num1", num1, "is greater than", "num2", num2)
elif num2 > num1:
    print("num2", "(",num2,")", "is greater than", "num1","(", num1,")")
else:
    print("num1", num1, "is equal to", "num2", num2)
```
